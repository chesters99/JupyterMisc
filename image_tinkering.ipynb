```python
from scipy.ndimage import convolve
hfilter = np.array(
    [[-1, -1, -1],
     [ 2,  2,  2],
     [-1, -1, -1]])

vfilter = np.array(
    [[-1,  2, -1],
     [-1,  2, -1],
     [-1,  2, -1]])

#Then apply it onto the image
filtered_image = convolve(gpic, vfilter)

#Show the result
plt.imshow(filtered_image)
plt.show()
```
